```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.io import wavfile
```

```python
all_audio_files = glob.glob("../microphone-sampling/NewTestingSamples/*/*")
light_rain_files = glob.glob("../microphone-sampling/NewTestingSamples/light_rain/*")
medium_rain_files = glob.glob("../microphone-sampling/NewTestingSamples/medium_rain/*")
heavy_rain_files = glob.glob("../microphone-sampling/NewTestingSamples/heavy_rain/*")
hail_files = glob.glob("../microphone-sampling/NewTestingSamples/hail/*")
heavy_hail_files = glob.glob("../microphone-sampling/NewTestingSamples/hail_heavy/*")
snow_files = glob.glob("../microphone-sampling/NewTestingSamples/snow_simulated/*")
whitenoise_files = glob.glob("../microphone-sampling/NewTestingSamples/whitenoise/*")
```

```python
def plot_spectrogram(wav_file):
    # Read the .wav file
    sample_rate, data = wavfile.read(wav_file)

    # If the data is stereo, take only one channel
    if len(data.shape) == 2:
        data = data[:, 0]

    # Plot the spectrogram
    plt.specgram(data, Fs=sample_rate, NFFT=1024, noverlap=512, cmap='inferno')
    plt.xlabel('Time [s]')
    plt.ylabel('Frequency [Hz]')
    plt.title('Spectrogram')
    plt.colorbar(label='Intensity [dB]')
    plt.show()
```

```python
for file in light_rain_files[0:5]:
    plot_spectrogram(file)
```
